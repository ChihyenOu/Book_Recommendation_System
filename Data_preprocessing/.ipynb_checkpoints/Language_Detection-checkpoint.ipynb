# Detection Language based on Item Title

## 1. Import item Dataset and Add a Column for Language

```python
import pandas as pd
items_dataset = pd.read_csv('../Original_dataset/items.csv',sep='|')

items_dataset ['language'] = ''
```

## 2. Language Detection

```python
!pip install langdetect
```

```python
from langdetect import DetectorFactory, detect
from langdetect.lang_detect_exception import LangDetectException
DetectorFactory.seed = 0
for i in range (len(items_dataset.index)):
    try:
        language = detect(items_dataset.loc[i].at['title'])
    except LangDetectException:
        language = 'unknown'
    items_dataset.at[i,'language'] = language
```

### 2.1 Language Distribution

```python
lang_type = items_dataset.groupby('language',sort = False).agg({'itemID': 'count'}).reset_index()
lang_type.sort_values(by=['itemID'], ascending=False, inplace=True, ignore_index= True)
print(lang_type)
```

### 2.2 Only Keep the 7 Most Common Language

```python
threshold = 1000
m = lang_type['itemID'] < threshold
lang_type1 = lang_type[~m].append(lang_type.loc[m, ['itemID']].sum(), ignore_index=True)
lang_type1.at[7,'language'] = 'other langs'
print(lang_type1)
```

### 2.3 Most Common Language Distribution

```python
import matplotlib.pyplot as plt
lang_labels = lang_type1.language
lang_sizes = lang_type1.itemID
fig1, ax1 = plt.subplots()
ax1.pie(lang_sizes, labels=lang_labels, autopct='%1.1f%%', startangle=90)
plt.title('Language distribution in items dataset')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.set_size_inches(7,7)
fig.gca().add_artist(centre_circle)
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
